```python
%pylab inline

Gamma1 = genfromtxt("gamma1.dat")
Gamma2 = genfromtxt("gamma2.dat")

fGamma1 = fft.fft2(Gamma1)
fGamma2 = fft.fft2(Gamma2)

k1 = fft.fftfreq(Gamma1.shape[0])
k2 = fft.fftfreq(Gamma2.shape[0])

d1 = zeros(fGamma1.shape)
d2 = zeros(fGamma2.shape)

for i in range(fGamma1.shape[0]):
    for j in range(fGamma1.shape[1]):
        if (i==0 and j==0):
            d1[i,j] = 0.0 
            d2[i,j] = 0.0
        else:
            d1[i,j] = (1./(k1[i]**2+k2[j]**2)) * (k1[i]*k1[i] - k2[j]*k2[j])
            d2[i,j] = (1./(k1[i]**2+k2[j]**2)) * 2.*k1[i]*k2[j]
                    
kappa = d1*fGamma1 + d2*fGamma2

ikappa = fft.ifft2(kappa)

matshow(real(ikappa))
colorbar()
show

matshow(imag(ikappa))
colorbar()
show
```

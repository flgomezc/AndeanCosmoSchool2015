```python
Rockstar128 = loadtxt('Rockstar_Halo_Catalog_128.dat', usecols=[2])
Rockstar256 = loadtxt('Rockstar_Halo_Catalog_256.dat', usecols=[2])

AmigaHF128 = loadtxt('amiga_output_128.z0.000.AHF_halos', usecols=[3])
AmigaHF256 = loadtxt('amiga_output_256.z0.000.AHF_halos', usecols=[3])
```

```python
Bins = 20
MassScale = logspace(10,15,Bins)
```

```python
V = 60.**3 # 60 Mpc^3 h^-1
```

```python
MF_Rock_128 = histogram(Rockstar128,bins=MassScale)[0]/V
MF_Rock_256 = histogram(Rockstar256,bins=MassScale)[0]/V

MF_Amig_128 = histogram(AmigaHF128,bins=MassScale)[0]/V
MF_Amig_256 = histogram(AmigaHF256,bins=MassScale)[0]/V

Mbins = MassScale = logspace(10,15,Bins-1)
```

```python
fig=figure(figsize=(10,6))

plot(Mbins,MF_Rock_128,label="Rockstar128")
plot(Mbins,MF_Rock_256,label="Rockstar256")
plot(Mbins,MF_Amig_128,label="Amiga128")
plot(Mbins,MF_Amig_256,label="Amiga256")

xscale('log')
yscale('log')
ylim(0.000001,1)
legend(loc=0)
```
